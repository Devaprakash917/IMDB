#31.Outlier Detection

```python
Q1=data['Rating'].quantile(0.25)
Q3=data['Rating'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-0.5*IQR
upper_bound=Q3+1.5*IQR
outliers=data[(data['Rating']<lower_bound) | (data['Rating']>upper_bound)]
print(pd.DataFrame(outliers))
print(len(outliers))
```

```python
print("Dataset before outliers : ",data['Rating'].count())
data=data[(data['Rating']>=lower_bound) & (data['Rating']<=upper_bound)]
print("Dataset after outliers : ",data['Rating'].count())
```

```python
data[['Rating']].boxplot(figsize=(10,6))
plt.xticks(rotation=90)
plt.show()
```

#Detect ouliers using isolationforest

```python
from sklearn.ensemble import IsolationForest
num_df=data.select_dtypes(include=['number'])
outliers=IsolationForest(contamination=0.05,random_state=42)
pred=outliers.fit_predict(num_df)
data['outlier_flag']=pred   #-1 is outlier and +1 is normal
```

```python
data
```

```python
outliers=data[data['outlier_flag']==-1]
```

```python
print(outliers) #only outliers
```
